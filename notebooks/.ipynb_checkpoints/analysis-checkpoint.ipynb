### Importing Packages

```python
import os
import cv2 as cv
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

### Image Directory Validation and Image Counts

```python
image_directory = './images/preprocessing'
if not os.path.exists(image_directory):
    raise Exception(f'Directory {image_directory} not dound')

image_count=0
imageFile = []
for file in os.listdir(image_directory):
    if (os.path.isfile(os.path.join(image_directory,file)) and os.path.splitext(file)[1]=='.jpg'):
        #print(file)
        imageFile.append(file)
        image_count += 1
print(f'Total Number of .jpg file in directory {image_directory} is {image_count}')
```

### Sorting Image List

```python
imageFile = sorted(imageFile, key= lambda x: int(x.split('.')[0].replace("Picture", "").replace('_','')) ,reverse=False )
print(imageFile)
```

```python
for file in  imageFile:
    imagepath = os.path.join(image_directory,file)
    image = cv.imread(imagepath)
    print(f'image {file} shape {image.shape}')
```

### Image Ploting Function

```python
def plot_2image(image_paths, dirr=image_directory):
    numberofImage=len(image_paths)
    fig, axes = plt.subplots(nrows=1, ncols=numberofImage)
    imagepaths = list(map(lambda x: os.path.join(dirr, x), image_paths))
    for file, ax in zip(imagepaths, axes.ravel()):
        img = mpimg.imread(file)
        ax.imshow(img,cmap='gray')
        ax.axis('off')
        ax.set_title(os.path.basename(file))
    fig.tight_layout()
```

```python
plot_2image(imageFile[2:4])
```

```python
img1 = cv.imread(os.path.join(image_directory,imageFile[2]))
img2 = cv.imread(os.path.join(image_directory,imageFile[3]))
print(f'image1 Shape {img1.shape}')
print(f'image2 Shape {img2.shape}')
diff = cv.absdiff(img1, img2)
mask = cv.cvtColor(diff, cv.COLOR_BGR2GRAY)
th = 10
imask =  mask>th
canvas = np.zeros_like(img2, np.uint8)
canvas[imask] = img2[imask]
cv.imwrite("./images/result/diff/diff_2_3.png", canvas)
cv.imwrite("./images/result/diff/mask_2_3.png", mask)
```

```python
output_dir='./images/Threshold2'
for file in  imageFile:
    imagepath = os.path.join(image_directory,file)
    image = cv.imread(imagepath, cv.IMREAD_GRAYSCALE)
    #thresh = cv.adaptiveThreshold(image, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY, 11,2)
    RET,thresh = cv.threshold(image,210,255,cv.THRESH_BINARY)
    print(f'processing image {file} shape {image.shape}')
    output_image=os.path.join(output_dir,file)
    cv.imwrite( output_image , thresh)
```
